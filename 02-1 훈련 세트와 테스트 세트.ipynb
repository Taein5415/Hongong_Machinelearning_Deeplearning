#훈련 세트와 테스트 세트
##지도 학습과 비지도 학습
데이터와 정답 : '입력', '타깃'이라고 부름  
데이터와 정답을 합쳐 '훈련 데이터'라고 부름

  
##훈련 세트와 테스트 세트
테스트 세트 : 평가에 사용하는 데이터  
훈련 세트 : 훈련에 사용되는 데이터

```python
from sklearn.neighbors import KNeighborsClassifier

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,          ##연어의 길이와 무게
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]      ##빙어의 무게와 길이
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

length = bream_length + smelt_length
weight = bream_weight + smelt_weight

fish_data = [[l,w] for l,w in zip(length, weight)]
fish_target = [1]*35 + [0]*14

train_input = fish_data[:35]    ##훈련 세트
train_target = fish_target[:35]

test_input = fish_data[35:]     ##테스트세트
test_target = fish_target[35:]

kn  = KNeighborsClassifier()
kn = kn.fit(train_input, train_target)

kn.score(test_input, test_target)
```

##넘파이
파이썬의 대표적인 배열 라이브러리  
고차원의 배열을 손 쉽게 만들고 조작할 수 있다


```python
import numpy as np
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print(input_arr)
```

shape 속성을 사용하여 샘플 수와 특성의 수를 확인 할 수 있다.

```python
print(input_arr.shape)
```

```python
index = np.arange(49)      ##0부터 N-1까지의 정수가 저장된 배열을 생성
np.random.shuffle(index)    ##셔플로 index 배열을 섞어줌

train_input = input_arr[index[:35]]       ##섞여진 index 배열을 사용해 배열슬라이싱으로 값 저장
train_target = target_arr[index[:35]]

test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0],train_input[:,1])    ##train_input의 0행과 1행을 사용
plt.scatter(test_input[:,0], test_input[:,1])     ##test_input의 0행과 1행을 사용

plt.xlabel('length')
plt.ylabel('weight')

plt.show()     ##파란색 : 훈령 세트, 주황색 : 테스트세트(도미와 빙어가 섞여있는 형태)
```

```python
kn = kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```
